```python
Y_train = dfy.values
Y_train = Y_train.reshape(-1)
Y_train.shape
```

```python
def dist(x1, x2):
  return np.sqrt(sum((x1-x2)**2))

def knn(X, Y, query_point, k = 20):
  vals = []
  m = X.shape[0]

  for i in range(m):
    d = dist(query_point, X[i])
    vals.append((d, Y[i]))

  vals = sorted(vals)
  vals = vals[:k]

  vals = np.array(vals)

  new_vals = np.unique(vals[:, 1], return_counts=True)

  max_freq = new_vals[1].argmax()
  pred = int(new_vals[0][max_freq])
  return pred



  # zero = 0
  # one = 0

  # for i in range(k):
  #   if vals[i][1] == 0:
  #     zero += 1
  #   else:
  #     one += 1
  
  # if(zero > one):
  #   return 0.0
  # else:
  #   return 1.0 
```

```python
dt = pd.read_csv("./Diabetes_Xtest.csv")
X_test = dt.values
```

```python
X_test.shape
```

```python
test_val = []
m = X_test.shape[0]

for i in range(m):
  ans = knn(X_train, Y_train, X_test[i])
  test_val.append(ans)
```

```python
out = np.array(test_val)
DF = pd.DataFrame(out)
```
