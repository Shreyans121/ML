```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=1.8, random_state=11)

print(X.shape)
Y = Y.reshape(-1, 1)
print(Y.shape)
```

```python
print(X)
```

```python
#Normalize
X = (X-X.mean())/X.std()

#Visualize
plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))
print(ones.size)
```

```python
X_ = np.hstack((X, ones))
print(X_.shape)
```

```python
def predict(X, theta):
    return np.dot(X, theta)

def getTheta(X, Y):
    firstPart = np.dot(X.T, X)
    secondPart = np.dot(X.T, Y)

    theta = np.dot(np.linalg.pinv(firstPart), secondPart)
    return theta
```

```python
print(predict(X_, theta).shape)
print(X.shape)
```

## Theta

```python
theta = getTheta(X_, Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, predict(X_, theta), color="red")
plt.title("Normalized Data")
plt.show()
```
