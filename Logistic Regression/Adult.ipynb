# Cleaning and Encoding Data

```python
df_clean = df.replace('?', np.nan)
df_clean = df_clean.fillna(df_clean.median())
df_clean.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col_enc = ['workclass', 'education','marital-status','occupation','relationship', 'race', 'sex', 'native-country', 'salary']
for i in col_enc:
  df_clean[i] = le.fit_transform(df_clean[i].astype(str))
```

# Visualization

```python
import seaborn as sns
plt.figure(figsize=(20,20))
sns.heatmap(df_clean.corr(),annot=True)
plt.show()
```

```python
columns_to_drop = ['workclass','fnlwgt','education','occupation','race','native-country']
df = df_clean.drop(columns_to_drop,axis=1)
```

```python
df.info()
```

```python
X = df.values[:,:-1]
Y = df.values[:,-1]
```

```python
X.shape,Y.shape
```

```python
u = np.mean(X, axis=0)
std = np.std(X, axis=0)
X = (X-u)/std
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X.shape)
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
def sigmoid(x):
  return 1.0/(1.0 + np.exp(-x))

def hypothesis(X,theta):
  return sigmoid(np.dot(X,theta))

def error(X,y,theta):
  hi = hypothesis(X,theta)
  e = -1*np.mean((y*np.log(hi) + ((1-y)*np.log(1-hi))))
  return e

def gradient(X,y,theta):
  hi = hypothesis(X,theta)
  grad = -(np.dot(X.T,(y-hi)))
  m = X.shape[0]
  return grad/m

def gradientDescent(X_,y,lr = 0.1, epochs = 300):
  y = y.reshape((-1,1))
  n = X_.shape[1]
  
  theta = np.zeros((n,1))
  error_list = []

  for i in range(epochs):
    error_list.append(error(X_,y,theta))
    grad = gradient(X_,y,theta)
    theta = theta - lr*grad
  return theta,error_list
```

```python
theta,error_list = gradientDescent(X_train,y_train, lr=0.1,epochs=500)
```

```python
plt.plot(error_list)
plt.xlabel("Epochs")
plt.ylabel("Loss/Error")
plt.show()
```

# Predictions

```python
def predict(X_,theta):
  h = hypothesis(X_,theta)
  output = np.zeros(h.shape)
  output[h>=0.5] = 1
  output = output.astype('int')
  return output
```

```python
XT_pred = predict(X_test,theta)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

CM = confusion_matrix(XT_pred, y_test, labels=[1,0])

TP=CM[0][0]
FP=CM[0][1]
FN=CM[1][0]
TN=CM[1][1]

ACC = (TP+TN)/(TP+TN+FP+FN)
print('Accuracy is : \n', ACC)
print()

Rec = TP/(TP+FN)
print('Recall is : \n', Rec)
print()

Prec = TP/(TP+FP)
print('Precision is : \n', Prec)
print()

F1 = 2 * ((Prec * Rec)/(Prec + Rec))
print('F1 score is : \n', F1)
```

```python
plt.figure(figsize=(12,9))
plt.ticklabel_format(useOffset=False, style='plain')
plt.title("Confusion Matrix")
sns.heatmap(CM,annot=True,cmap=plt.cm.Blues)
plt.plot()
```

```python
def probabilities(X_,theta):
  h = hypothesis(X_,theta)
  return h
```

```python
prob_vector = probabilities(X,theta)
```

# ROC Curve

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(Y, prob_vector)
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(fpr, tpr, s=100, alpha=0.5, color="blue", label="ROC Curve")
plt.title("ROC Curve", fontsize=20)
plt.xlabel("False Positive Rate", fontsize=16)
plt.ylabel("True Positive Rate", fontsize=16)
plt.legend()
plt.show()
```
