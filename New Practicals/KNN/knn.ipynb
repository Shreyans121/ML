```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('xdata.csv')
Y = pd.read_csv('ydata.csv')
```

```python
x = X.values
y = Y.values
x = x[:,1:]
y = y[:,1:].reshape((-1,))
```

```python
x.shape,y.shape
```

```python
plt.style.use('seaborn')
plt.scatter(x[:,0],x[:,1],c = y)
plt.show()
```

```python
query_x = np.array([2,3])
plt.scatter(query_x[0],query_x[1],c='red')
plt.scatter(x[:,0],x[:,1],c = y)
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))

def knn(x,y,query,k=5):
    dist_val = []
    m = x.shape[0]
    for i in range(m):
        d = dist(query,x[i])
        dist_val.append((d,y[i]))
    dist_val = sorted(dist_val)
    dist_val = dist_val[:k]
    vals = np.array(dist_val)
    new_vals = np.unique(vals[:,1],return_counts=True)
    index = np.argmax(new_vals[1])
    return new_vals[0][index]
```

```python
pred = knn(x,y,query_x)
print(int(pred))
```
